```python
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import model_selection, datasets, linear_model, metrics
```

```python
train_data = pd.read_csv('orange_small_churn_data.txt')
labels = pd.read_csv('orange_small_churn_labels.txt', names=['target'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_data, labels, test_size=0.25, random_state=7, stratify=labels)
```

```python
X_test.to_csv("Hold_out_X.csv", index = False)
y_test.to_csv("Hold_out_y.csv", index = False)
X_train.to_csv('X_train.csv', index = False)
y_train.to_csv('y_train.csv', index = False)
```
